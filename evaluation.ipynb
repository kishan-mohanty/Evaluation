```python
import requests
```

```python
conferences_details=requests.get("https://o136z8hk40.execute-api.us-east-1.amazonaws.com/dev/get-list-of-conferences")
```

```python
conference_details_json=conferences_details.json()
```

```python
for conference in conference_details_json["paid"]:
    print("**************"+str(conference["confName"])+"****************")
    print("\n")
    v=str(conference["confName"])+" at  "+str(conference["venue"])+"\n"+"from "+str(conference["confStartDate"])+" to "+str(conference["confEndDate"])
    print(v)
    k="keyword:"+str(conference["keywordSupport"])
    print(k)
    m="Conference Details:"+"\n"+"Conference id is"+str(conference["conference_id"])+"\n"+"REgistration_URL:"+str(conference["confRegUrl"])+"\n"+"Entry:"+str(conference["entryType"])
    print(m)
    print("\n")
    r="For more details you can visit:"+"\n"+"website:"+str(conference["confUrl"])+"\n"+"twitter:"+str(conference["twitter_handle"])
    print(r)

    print("\n")
    print("\n")
```
